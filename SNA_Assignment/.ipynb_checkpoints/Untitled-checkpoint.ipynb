```python
weightDF = pd.read_excel("MetroDataWeight.xlsx")
```

```python
weightDF.head()
```

```python
data = weightDF.loc[:,'From_Station':].as_matrix()
Madrid_tube = nx.Graph()
Madrid_tube.add_nodes_from(list(range(0,369)))
Madrid_tube.add_weighted_edges_from(data)
```

```python
graph_dict = dict(list(zip(['Madrid_tube'],
                      [Madrid_tube])))
for key in graph_dict:
    print(key)
    print(nx.info(graph_dict[key]))
    print()
```

```python
fig, ax = plt.subplots(nrows=2,figsize=(10,20))
nx.draw(Madrid_tube,node_size=20,edge_cmap=cm.get_cmap('rainbow'),ax=ax[0])
```

```python
pos_dict = {}
pos_dict[1102] = list(nodesDF.loc[nodesDF.StationID == 1102,['Longitude','Latitude']].values)
pos_dict
```

```python
pos_dict = {}
for key in nodesDF.StationID:
    pos_dict[key] = list(nodesDF.loc[nodesDF.StationID == key,['Longitude','Latitude']].values[0])
print(pos_dict)
plt.figure(figsize=(30,30))
nx.draw(Madrid_tube,pos=pos_dict,node_size=20)
plt.show()
```

```python
G_tube_connected = [G for G in nx.subgraph(Madrid_tube) if G.connected_components()>1][0]

connected_graphs = [G_tube_connected]

for G in connected_graphs:
    print(nx.diameter(G))
```
